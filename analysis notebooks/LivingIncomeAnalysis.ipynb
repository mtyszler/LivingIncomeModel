# 1. Calculating the total income

## 1.1 Introduction

We calculate the total household income as:

_Total Household Income = (net cocoa income)/(income share from cocoa)_

In other words, I calculate the net cocoa income, and extrapolate the household income by using the information on the share of total income coming from cocoa

Te _net cocoa income_ is calculated as:

_net cocoa income = (total production (kg/ha) * price (USD/kg) - input costs (USD/ha) - labor costs (USD/ha))* Productive land (ha)_

To calculate input costs and labor costs I need to go over the different input types and production activities

## 1.2 Calculating income components

The detailed cocoa production features are built over a few production activities, types of labor and inputs. I will create lists with those activities, types of labor and inputs for further iteration in this notebook.

```python
production_activities = [
    'landclear',
    'landprep',
    'planting',
    'gfert',
    'lfert',
    'manurecomp',
    'herb',
    'pest',
    'fung',
    'weed',
    'pruning',
    'harvest',
    'podbreak',
    'ferment',
    'transport',
    'dry'
]

types_of_labor = [
    'hi_n_ha',
    'hh_n_ha',
    'com_n_ha',
    'lab_n_ha',
    'hi_lab_usdha'
]

inputs = [
    'gfert',
    'lfert',
    'herb',
    'pest',
    'fung'
]
```

Let's first check if these variables have missing values. I look at the `hi_lab_usdha` features, since this is directly relevant for the calculations:

```python
data.filter(regex = 'hi_lab_usdha').describe()
```

As expected these features have missing values, in different levels. As the previous studies explain, this happens when farmers know they did some activity, but did not know the exact costs. 

Let's verify:

```python
data.loc[data['cocoa_landclear_hi_lab_usdha'].isnull(),'cocoa_landclear_yn'].value_counts()
```

My strategy will be to:

- enforce 0 if activity was not done
- if the activity was done but it is missing, fill the missing values with the median values (per ha) of the farmers that did the activity and could specify the costs. I will do this per analytical group.

```python
def fill_na_by_group(df, yn_feature, feature, group_feature, strategy = 'median'):
    '''
    fills missing values in given feature with the group median (default) or mean value
        
    
    INPUT:
        df: panda DataFrame
        yn_feature: colname to be used to slice the dataframe into doing (yes) or not doing (no) an activity
        feature: colname to be filled
        group_feature: feature identifying the different groups
        strategy = 'median' or 'mean'
        
    OUTPUT:
        df: modified DataFrame
    '''
    
    # first enforce 0 if yn_feature = 'no'
    df.loc[df[yn_feature]=='no', feature] = 0

    # get a list of groups
    groups = df[group_feature].unique()
    
    # for each group, filter on yn_feature = yes and fill in median
    for group in groups:
        
        if strategy == 'median':
            fill_value = df.loc[(df[group_feature] == group) & (df[yn_feature] == 'yes'), feature ].median()
        elif strategy == 'mean':
            fill_value = df.loc[(df[group_feature] == group) & (df[yn_feature] == 'yes'), feature ].mean()
        else:
            raise SyntaxError('Strategy unknown')
        
        df.loc[(df[group_feature] == group) & (df[yn_feature] == 'yes'), feature ] = df.loc[(df[group_feature] == group) & (df[yn_feature] == 'yes'), feature ].fillna(fill_value)
       
    
    # return modified DataFrame
    return df
```

```python
# iterate through all types of labor
for labor_type in types_of_labor:
    
    # find all columns with that labor type
    full_list = list(data.filter(regex = labor_type))
    
    # apply the fill_na_by_group to each of these columns
    for col in full_list:
        yn_feature = col.replace(labor_type,'yn')
        data = fill_na_by_group(data, yn_feature, col, 'Analytical Group')
```

## 1.3 Value adjustments over time

Since a main goal of this analysis is to compare total incomes with the Living Income Benchmark value, they need to be comparable in purchasing power. The Living Income Benchmark values refer to Q1 2018 (Ghana) and Q2 2018 (CÃ´te d'Ivoire). This data, on the other hand, refers to Q1 2016. Therefore I will use a [CPI correction factor](http://data.imf.org/regular.aspx?key=61545849) to bring the survey data from 2016 to 2018 values

```python
# CPI Correction

# Ghana CPI in 2018 Q1: 247.05
# Ghana CPI in 2016 Q1: 197.77
# Living Income  CEDI to USD = 1/4.45 // March 1st 2018
# Survey data CEDI to USD = 0.26116 // Jan 15th, 2016
correction_factor_ghana = (1/0.26116)/197.77*247.05/4.45


# CdI CPI in 2018 Q2: 112.92
# CdI CPI in 2016 Q1: 111.35
# Living Income CFA to USD = 1/576.81 // August 2018
# Survey data CFA to USD = 0.00166 // Jan 16th, 2016
correction_factor_cdi = (1/0.00166)/111.35*112.92/576.81
```

Apply the correction factor all variables in USD:

```python
usd_list = data.filter(regex = "usd").columns
usd_list = usd_list.append(data.filter(regex = "USD").columns)
```

```python
data.loc[data['country'] == 'Ghana', usd_list] = data.loc[data['country'] == 'Ghana', usd_list]*correction_factor_ghana
data.loc[data['country'] == 'CdI', usd_list] = data.loc[data['country'] == 'CdI', usd_list]*correction_factor_cdi
```

## 3.2 Investigate Missing values more in detail

I'll create a function to compute the number of missing values per variable, as a share of the total

```python
def missing_per_variable(data, describe = True):
    '''
    computes the number of missing values per variable
        
    INPUT:
        data: panda DataFrame
        describe: print a decription table
        
    OUTPUT:
        share_missing: panda Series with the share of missing values per variables
    '''
    share_missing = data.isnull().sum()/len(data.index)*100
    
    if describe:
        print(share_missing.describe(percentiles = [.25, .5, .75, .90, .95, .99],))
    
    return share_missing
```

I'll also create a function that plots the missing values per variable:

```python
def plot_missing(share_missing, cutoff = 5):
    '''
    plots the share of missing value per variable
        
    INPUT:
        share_missing: panda series created with missing_per_variable()
        cutoff: minimum share (0-100) to be included in the plot
        
    OUTPUT:
        nothing (plots the graph on the notebook)
    '''
    
    share_missing[share_missing>cutoff].sort_values().plot.bar(
        figsize=(15,10), 
        color=(0.2, 0.4, 0.6, 0.6)
    )
    plt.title("Ordered barplot of missing values per feature")
    plt.ylabel("Share of missing values (%)")
    plt.show()
```

Let's create a supporting function

```python
def impose_yn_logic(data, data_properties, main_feature, features_to_clean_main_stub = '', yes_no_label = 'lblYES_NO', verbose = 1):
    '''
    Imposes yes/no logic in the data. If a main question is `no` all children should also be `no`
        
    INPUT:
        data: panda DataFrame
        data_properties: StataRead dictionary object created with pd.read_stata(filename, iterator = True)
        main_feature: yes/no parent feature
        features_to_clean_main_stub (optional): main stub to find the children features. If not provided the main stub is 
            derived from the main_feature, removing `_yn` string
        yes_no_label: (stata) label that maps YES/NO questions
        verbose: if greater than 0, prints informative messages
        
    OUTPUT:
        data: modified panda DataFrame
    '''
    
        
    #if no main stub is passed, find the main stub, excluding the "yn"
    if features_to_clean_main_stub == '':
        main_stub = main_feature.replace("_yn","")
    else:
        main_stub = features_to_clean_main_stub
    
    # get a list of all other variable with the same stub
    features_to_clean = list(data.filter(regex = main_stub))
    try:
        features_to_clean.remove(main_feature)
    except:
        ;
    
    if verbose >=1:
        print('')
        print("Based on {}, clean:".format(main_feature))
        print(features_to_clean)
    
    
    # for each feature in the list, if the feature is a yes/no question, impose 'no' if main is 'no'
    for this_feature, this_label in zip(data_properties.varlist, data_properties.lbllist):
        if (this_feature in features_to_clean) & (this_label == yes_no_label):
            data.loc[data[main_feature]=='no', this_feature] = 'no'
            
    return data
    
```

```python
for main_question in yes_no_questions:
    
    # for each feature in the list, if the feature is a yes/no question, impose 'no' if main is 'no'
    data = impose_yn_logic(data, data_properties, main_question)
    
```

Before finishing, I'll replace the binary categories by their numeric values (0 or 1)

```python
def convert_tocodes(data, data_properties, label_name):
    '''
    Convert categories to numeric values, according to the stata dictionaty
        
    INPUT:
        data: panda DataFrame
        data_properties: StataRead dictionary object created with pd.read_stata(filename, iterator = True)
        label_name: (stata) label that maps categories in that question
        
    OUTPUT:
        data: modified panda DataFrame
    '''
    
    # make a dataframe
    data_and_label = pd.DataFrame({
        "variable label": data_properties.varlist,
        "value label": data_properties.lbllist})
    
    # filter on the selected label
    data_and_label = data_and_label[data_and_label['value label']==label_name]

    # convert all features that use that label
    for this_column in list(data.columns):

        if this_column in list(data_and_label['variable label']):
            data[this_column] = data[this_column].cat.codes
            data.loc[data[this_column] == -1, this_column] = np.nan
            
    return data
    
```
